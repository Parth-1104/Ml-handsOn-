```python
import pandas as pd
```

```python
df=pd.read_csv('/content/air-quality-india.csv')
```

```python
df.sample(10)
df.isnull().sum()
```

```python
df['Timestamp'][0]
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['Date'] = df['Timestamp'].dt.date
```

```python
daily_avg = df.groupby('Date')['PM2.5'].mean().reset_index()
daily_avg.columns = ['Date', 'PM2_5_Avg']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 6))
sns.lineplot(data=daily_avg, x='Date', y='PM2_5_Avg', label='Daily Average PM2.5')
plt.title('Daily Average PM2.5 Concentration Over Time')
plt.xlabel('Date')
plt.ylabel('PM2.5 (µg/m³)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
threshold = 120
high_pm = daily_avg[daily_avg['PM2_5_Avg'] > threshold]

plt.figure(figsize=(14, 6))
sns.lineplot(data=daily_avg, x='Date', y='PM2_5_Avg', label='Daily Average PM2.5')
plt.scatter(high_pm['Date'], high_pm['PM2_5_Avg'], color='red', label='High PM2.5 Days')
plt.axhline(threshold, color='orange', linestyle='--', label=f'Threshold: {threshold}')
plt.title('High PM2.5 Events Highlighted')
plt.xlabel('Date')
plt.ylabel('PM2.5 (µg/m³)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Do temperature and humidity rise/fall together?
Is pressure lower during hotter periods?
Any seasonal spikes or consistent cyclic behavior?

```python
params = ['meantemp', 'humidity', 'wind_speed', 'meanpressure']
plt.figure(figsize=(14, 10))

for i, param in enumerate(params):
    plt.subplot(2, 2, i+1)
    sns.histplot(df2[param], kde=True, bins=30)
    plt.title(f"Distribution of {param}")

plt.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(14, 5))

ax1.set_title("Temperature vs Humidity Over Time")
ax1.set_xlabel("Date")
ax1.set_ylabel("Mean Temperature (°C)", color='red')
ax1.plot(df2.index, df2['meantemp'], color='red', label='Temp')
ax1.tick_params(axis='y', labelcolor='red')

ax2 = ax1.twinx()
ax2.set_ylabel("Humidity (%)", color='blue')
ax2.plot(df2.index, df2['humidity'], color='blue', label='Humidity')
ax2.tick_params(axis='y', labelcolor='blue')

fig.tight_layout()
plt.show()
```

```python
df_rolling = df2[['meantemp']].rolling(window=30).mean()

plt.figure(figsize=(14, 5))
plt.plot(df2['meantemp'], label='Original', alpha=0.5)
plt.plot(df_rolling, label='30-Day Rolling Mean', color='black')
plt.title('Trend in Temperature with Rolling Average')
plt.legend()
plt.show()
```

```python
import pandas as pd


df3 = pd.read_csv('/content/air-quality-india.csv', parse_dates=['Timestamp'], index_col='Timestamp')


print(df3.head())
```

```python
import matplotlib.pyplot as plt

df3['PM2.5'].plot(figsize=(14, 6), title='Hourly PM2.5 Levels Over Time in India')
plt.xlabel('Date')
plt.ylabel('PM2.5 (µg/m³)')
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(df3['PM2.5'].dropna())
print('ADF Statistic:', result[0])
print('p-value:', result[1])
```

```python
from statsmodels.tsa.stattools import kpss

kpss_stat, p_value, _, _ = kpss(df3['PM2.5'].dropna(), regression='c')
print('KPSS Statistic:', kpss_stat)
print('p-value:', p_value)
```
